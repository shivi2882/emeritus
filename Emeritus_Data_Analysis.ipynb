### Average cost per model per year

```python
avgg = df.groupby(['Year', 'Model Number'])['Material Cost'].mean().reset_index()
```

```python
df.groupby(['Year', 'Model Number'])['Material Cost'].mean().reset_index() \
.sort_values(by='Material Cost', ascending=False).head(50)
```

```python
avg_table = avgg.pivot(index='Model Number', columns='Year', values='Material Cost')
```

### Expensive beds (based on repair costs) by Asset ID

```python
asset_costs = df.groupby('Asset ID')['Material Cost'].agg(
    total_cost='sum',
    average_cost='mean',
    number_of_repairs='count'
).reset_index()
```

### Annual costs per model per age

```python
df['Install Date'] = pd.to_datetime(df['Install Date'])
df['WO Date'] = pd.to_datetime(df['WO Date'])
df['Age (Years)'] = (df['WO Date'] - df['Install Date']).dt.days / 365.25
df['Age (Years)'] = df['Age (Years)'].apply(lambda x: max(x, 1))
df['Annual Material Cost'] = df['Material Cost'] / df['Age (Years)']
avgg_annual = df.groupby('Model Number')['Annual Material Cost'].mean().reset_index()
```
